# Question 2: What is TextBlob and how does it simplify common NLP tasks like sentiment analysis and translation?


**TextBlob** is a simple and user-friendly Python library built on top of **NLTK (Natural Language Toolkit)**. It provides an easy interface for performing common **Natural Language Processing (NLP)** tasks without requiring complex code.

TextBlob is especially suitable for beginners because it allows developers to perform various NLP operations with just a few lines of Python code.



# How TextBlob Simplifies Sentiment Analysis

**Sentiment Analysis** is the process of determining whether a piece of text expresses a **positive, negative, or neutral** opinion.

TextBlob automatically calculates two important values:

### 1. Polarity

- Range: **-1 to +1**
- **-1** → Negative sentiment
- **0** → Neutral sentiment
- **+1** → Positive sentiment

### 2. Subjectivity

- Range: **0 to 1**
- **0** → Objective (fact-based)
- **1** → Subjective (opinion-based)

### Example

**Text:**
```
"The customer service is excellent."
```

Output:

- Polarity = Positive
- Subjectivity = High

This makes it easy to analyze customer reviews, product feedback, and social media posts.



# How TextBlob Simplifies Translation

TextBlob also provides built-in support for **language translation**.

It can translate text from one language to another using translation services.

### Example

English:
```
Good Morning
```

Hindi:
```
सुप्रभात
```

This feature is useful for multilingual applications, chatbots, customer support systems, and international communication.

# Question 4: Describe the architecture and functioning of a Recurrent Neural Network (RNN).

# Architecture of a Recurrent Neural Network (RNN)

An RNN consists of the following main components:

### 1. Input Layer

The input layer receives sequential data one element at a time.

Examples:

- Words in a sentence
- Audio signals
- Stock prices
- Weather data

Each input at a time step is represented as **xₜ**.



### 2. Hidden Layer

The hidden layer is the core component of an RNN.

At every time step, it receives:

- The current input (**xₜ**)
- The hidden state from the previous time step (**hₜ₋₁**)

Using both pieces of information, it computes a new hidden state (**hₜ**). This hidden state acts as the network's memory, allowing information from earlier time steps to influence future predictions.



### 3. Recurrent Connection

The hidden layer has a feedback (recurrent) connection that passes the hidden state from one time step to the next.

This enables the network to remember previous information and capture dependencies in sequential data.


### 4. Output Layer

The output layer generates the prediction for each time step.

Depending on the application, the output may be:

- One output after processing the entire sequence (e.g., sentiment analysis)
- One output for every time step (e.g., language translation)



# Working of an RNN

The functioning of an RNN can be explained in the following steps:

### Step 1: Receive Sequential Input

The network receives the first input (**x₁**) and initializes the hidden state.



### Step 2: Compute Hidden State

The current input and the previous hidden state are combined to compute a new hidden state.

This hidden state stores information from previous inputs.



### Step 3: Generate Output

The hidden state is used to produce the output for the current time step.



### Step 4: Pass Memory Forward

The hidden state is passed to the next time step, allowing the network to remember previous information.



### Step 5: Repeat for Entire Sequence

The above process continues until all elements of the sequence have been processed.



# Question 5: What is the key difference between LSTM and GRU networks in NLP applications?



| Feature | LSTM | GRU |
|---------|------|-----|
| **Full Form** | Long Short-Term Memory | Gated Recurrent Unit |
| **Memory Cell** | Yes | No |
| **Number of Gates** | Three (Forget, Input, Output) | Two (Update, Reset) |
| **Hidden State** | Hidden state and cell state | Hidden state only |
| **Architecture** | More complex | Simpler |
| **Number of Parameters** | More | Fewer |
| **Training Speed** | Slower | Faster |
| **Memory Usage** | Higher | Lower |
| **Performance on Long Sequences** | Generally better | Good, but may be slightly less effective on very long sequences |
| **Best Use Cases** | Complex NLP tasks with long dependencies | Faster NLP applications and smaller datasets |



````python
# Question 6
# Perform Sentiment Analysis using TextBlob

# Import TextBlob
from textblob import TextBlob

# Input paragraph
text = """
I had a great experience using the new mobile banking app.
The interface is intuitive, and customer support was quick to resolve my issue.
However, the app did crash once during a transaction, which was frustrating.
"""

# Create a TextBlob object
blob = TextBlob(text)

# Perform sentiment analysis
sentiment = blob.sentiment

# Print the results
print("Sentiment Analysis Results")
print("-" * 35)

print("Polarity Score    :", sentiment.polarity)
print("Subjectivity Score:", sentiment.subjectivity)

# Interpret the polarity
if sentiment.polarity > 0:
    print("\nOverall Sentiment : Positive")
elif sentiment.polarity < 0:
    print("\nOverall Sentiment : Negative")
else:
    print("\nOverall Sentiment : Neutral")
````

````python
# Question 7
# Perform String Tokenization and Frequency Distribution using Python and NLTK

# Import required libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download required NLTK resources (Run only once)
nltk.download('punkt')
nltk.download('punkt_tab')

# Sample paragraph
text = """
Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.
"""

# Convert text to lowercase
text = text.lower()

# Tokenize the text into words
tokens = word_tokenize(text)

# Display word tokens
print("Word Tokens:\n")
print(tokens)

# Create frequency distribution
freq_dist = FreqDist(tokens)

# Display frequency distribution
print("\nWord Frequency Distribution:\n")
for word, frequency in freq_dist.items():
    print(f"{word} : {frequency}")

# Display the 10 most common words
print("\nTop 10 Most Frequent Words:\n")
print(freq_dist.most_common(10))
````

````python
# Question 8
# Build a basic LSTM model in Keras for text classification

# Import required libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# -----------------------------
# Dataset
# -----------------------------
texts = [
    "I love this project",                # Positive
    "This is an amazing experience",      # Positive
    "I hate waiting in line",             # Negative
    "This is the worst service",          # Negative
    "Absolutely fantastic!"               # Positive
]

labels = [1, 1, 0, 0, 1]

# -----------------------------
# Text Preprocessing
# -----------------------------

# Initialize tokenizer
tokenizer = Tokenizer()

# Build vocabulary
tokenizer.fit_on_texts(texts)

# Convert text to sequences
sequences = tokenizer.texts_to_sequences(texts)

# Pad sequences to equal length
max_length = max(len(seq) for seq in sequences)
X = pad_sequences(sequences, maxlen=max_length, padding="post")

# Convert labels to NumPy array
y = np.array(labels)

# Display tokenized sequences
print("Tokenized Sequences:\n")
print(X)

# -----------------------------
# Build LSTM Model
# -----------------------------

model = Sequential([
    Embedding(
        input_dim=len(tokenizer.word_index) + 1,
        output_dim=16
    ),

    LSTM(32),

    Dense(1, activation="sigmoid")
])

# Compile the model
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

# -----------------------------
# Train the Model
# -----------------------------

history = model.fit(
    X,
    y,
    epochs=20,
    batch_size=2,
    verbose=1
)

# -----------------------------
# Evaluate the Model
# -----------------------------

loss, accuracy = model.evaluate(X, y, verbose=0)

print("\nTraining Accuracy: {:.2f}%".format(accuracy * 100))
````

````python
# Question 9
# Build a simple NLP pipeline using spaCy:
# Tokenization, Lemmatization, and Named Entity Recognition (NER)

# Import spaCy
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Input paragraph
text = """
Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India's atomic energy program. He was the founding director of the
Tata Institute of Fundamental Research (TIFR) and was instrumental in establishing
the Atomic Energy Commission of India.
"""

# Process the text
doc = nlp(text)

# ----------------------------------------
# Tokenization
# ----------------------------------------

print("TOKENS\n")

for token in doc:
    print(token.text)

# ----------------------------------------
# Lemmatization
# ----------------------------------------

print("\n" + "="*50)
print("LEMMATIZATION\n")

for token in doc:
    print(f"{token.text:<20} --> {token.lemma_}")

# ----------------------------------------
# Named Entity Recognition (NER)
# ----------------------------------------

print("\n" + "="*50)
print("NAMED ENTITIES\n")

for entity in doc.ents:
    print(f"{entity.text:<50} {entity.label_}")
````

# Question 10: Mental Health Chatbot using LSTM/GRU and spaCy

## Problem Statement

````python
import spacy
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Sample dataset
texts = [
    "I feel very happy today",
    "I am feeling sad",
    "I have anxiety",
    "I feel stressed",
    "Hello",
    "Good morning"
]

# Intent labels
labels = [1, 2, 3, 4, 0, 0]

# Text preprocessing using spaCy
processed_texts = []

for sentence in texts:
    doc = nlp(sentence.lower())

    words = [
        token.lemma_
        for token in doc
        if not token.is_stop and not token.is_punct
    ]

    processed_texts.append(" ".join(words))

print("Processed Text:")
print(processed_texts)

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(processed_texts)

sequences = tokenizer.texts_to_sequences(processed_texts)

# Padding
max_length = max(len(seq) for seq in sequences)
X = pad_sequences(sequences, maxlen=max_length, padding="post")

y = np.array(labels)

# Build LSTM Model
model = Sequential([
    Embedding(input_dim=len(tokenizer.word_index) + 1,
              output_dim=16),

    LSTM(32),

    Dense(16, activation="relu"),

    Dense(5, activation="softmax")
])

# Compile Model
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# Train Model
history = model.fit(
    X,
    y,
    epochs=20,
    batch_size=2,
    verbose=1
)

# Evaluate Model
loss, accuracy = model.evaluate(X, y, verbose=0)

print("\nTraining Accuracy: {:.2f}%".format(accuracy * 100))
````
