1. What is the difference between interpreted and compiled languages?

 - Compiled languages (like C, C++, Java): The entire code is converted into machine code (binary) before running. This makes them faster but debugging is harder.
 - Interpreted languages (like Python, JavaScript): The code is executed line by line at runtime by an interpreter. They are slower than compiled languages but easier to write, debug, and platform-independent.

 - Python is an interpreted language.

2. What is exception handling in Python?
 - Exception handling is a way to handle runtime errors (like division by zero, file not found, etc.) so that the program does not crash suddenly.
 - We use try, except, else, finally blocks to manage these errors gracefully.

3. What is the purpose of the finally block in exception handling?
 - The finally block contains code that always executes, whether an exception occurs or not.
 - It is mainly used for cleanup tasks like closing files, releasing resources, or disconnecting databases.

7. How can you handle multiple exceptions in Python?
 - In Python, multiple exceptions can occur while executing a program. To prevent the program from crashing, we use multiple except blocks or handle multiple exceptions in a single except block.

- This allows us to execute different code for different types of errors.

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num

except ValueError:
    print("Invalid input! Please enter a valid number.")

except ZeroDivisionError:
    print("Cannot divide by zero.")

print("Program Ended")
```

8. What is the purpose of the with statement when handling files in Python?
 - The with statement is used in Python for safe and efficient file handling. It automatically manages resources by opening a file and ensuring that it is properly closed after use, even if an error occurs during file operations.

 - Without the with statement, we need to manually close the file using the close() method. Forgetting to close a file can lead to memory leaks and resource wastage.

```python
# Writing to a file

with open("sample.txt", "w") as file:
    file.write("Hello Python")   # File is automatically closed here

print("Data written successfully.")
```

11. What is memory management in Python?
 - Memory Management in Python is the process of allocating, using, and freeing memory efficiently while a program is running. Python automatically manages memory, so programmers do not need to manually allocate or deallocate memory like in languages such as C or C++.

 - Python uses a combination of reference counting and garbage collection to manage memory.

2. What are the basic steps involved in exception handling in Python?
 - Step 1: Write Risky Code inside the try Block

    - The code that may cause an error is placed inside the try block.

 - Step 2: Handle Errors using the except Block

   - If an exception occurs, Python immediately transfers control to the corresponding except block.

 - Step 3: Execute the else Block (Optional)

   - The else block runs only if no exception occurs in the try block.

 - Step 4: Execute the finally Block (Optional)

    - The finally block always executes whether an exception occurs or not. It is commonly used for cleanup operations such as closing files or database connections.

13. Why is memory management important in Python?
 - Importance of Memory Management in Python
   -  Efficient Use of Memory : Memory management ensures that only the required memory is allocated to objects and that unused memory is released.
   - Prevents Memory Leaks : A memory leak occurs when memory is occupied by unused objects and is not released properly. Python's garbage collector automatically removes unused objects, helping prevent memory leaks.
   - Improves Program Performance : Efficient memory usage allows programs to execute faster and reduces unnecessary memory consumption.
   - Prevents Program Crashes : If memory is not managed properly, a program may consume excessive memory and crash. Python automatically frees unused memory to reduce this risk.
   - Supports Large Applications : Large applications create thousands of objects during execution. Memory management ensures that these objects do not occupy memory unnecessarily.
   - Reduces Programmer Effort : Unlike languages such as C or C++, Python automatically handles memory allocation and deallocation.

14. What is the role of try and except in exception handling?
 - In Python, the try and except blocks are used to handle errors (exceptions) that may occur during program execution.

 - Instead of terminating the program when an error occurs, Python uses try and except to catch and handle the error gracefully, allowing the program to continue running.
 - Role of try Block:

    - The try block contains the code that may potentially generate an exception.

    - Python first executes the code inside the try block. If no error occurs, the code runs normally.

  - Role of except Block:

    - The except block is used to handle exceptions that occur in the try block.

    - If an error occurs, Python immediately stops executing the remaining code in the try block and transfers control to the matching except block.

15. How does Python's garbage collection system work?
 - Python's Garbage Collection (GC) system is responsible for automatically freeing memory occupied by objects that are no longer being used by the program.

 - The main purpose of garbage collection is to identify and remove unused objects, preventing memory leaks and improving memory efficiency.

 - Python uses two mechanisms for garbage collection:

    - Reference Counting :
        -  Reference counting is the primary memory management technique used by Python.
        - Every object in Python keeps track of how many variables are referring to it.

        - When the reference count becomes zero, Python automatically deletes the object and frees its memory.
    - Generational Garbage Collection:
        - Python's Garbage Collector detects and removes circular references.
        - It periodically checks objects and removes those that are no longer reachable by the program.
        

16. What is the purpose of the else block in exception handling?
 - In Python, the else block is used in exception handling to execute code only when no exception occurs in the try block.

 - The else block is optional and is placed after the except block(s). It helps separate the code that should run when the try block executes successfully from the code that handles errors.

```python
try:
    result = 10 / 2
except:
    print("Error")
else:
    print("Success:", result)   # Runs only if no error
```

```python
#18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?
import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler(
    "app.log",
    maxBytes=1024*1024,  # 1 MB
    backupCount=3
)

logging.basicConfig(
    handlers=[handler],
    level=logging.INFO
)

logging.info("Log rotation example.")
```
