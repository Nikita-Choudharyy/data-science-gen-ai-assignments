3. What are the different ways to define and call a function in Python?
 - In Python, functions can be defined and called in different ways depending on the requirement. Some common methods are:

    1. Function Without Parameters
       - A function that does not take any input values.
    2. Function With Parameters
       - A function that accepts values (parameters) when it is called.
    3. Function With Return Value
       - A function can return a result using the return statement.
    4. Function With Default Arguments
       - Default values are assigned to parameters.
    5. Keyword Arguments Function
       - Arguments are passed using parameter names.

```python
#Function Without Parameters
def greet():
    print("Welcome")

greet()

print('-'*20)

#Function With Parameters
def greet(name):
    print("Hello", name)

greet("Nikita")

print('-'*20)

#Function With Return Value
def square(num):
    return num ** 2

print(square(5))

print('-'*20)

#Function With Default Arguments
def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Nikita")

print('-'*20)

#Keyword Arguments Function
def student(name, age):
    print(name, age)

student(age=20, name="Nikita")
```

5. What are iterators in Python and how do they differ from iterables?

 - Iterable: An object that can return an iterator (implements __iter__()).
Examples: list, tuple, string, dict.
 - Iterator: An object that keeps track of state and produces the next value (implements __iter__() and __next__()). Once exhausted, it cannot be reused.

```python
# 1. Iterable
numbers = [10, 20, 30, 40, 50]   # This is an ITERABLE

# 2. Creating Iterator from Iterable
it = iter(numbers)   # This is an ITERATOR

print(type(numbers))  # <class 'list'>
print(type(it))       # <class 'list_iterator'>
```

6. Explain the concept of generators in Python and how they are defined.
 - Generators are special iterators created with functions containing the yield keyword. They produce values lazily (one at a time) and pause/resume execution.

   - Use yield instead of return
   - Generator expression: (x*x for x in range(10))

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count          # Pause here, save state
        count += 1

gen = count_up_to(3)
print(next(gen))  # 1
print(next(gen))  # 2
print(list(gen))  # [3]
```

7. What are the advantages of using generators over regular functions?

 - Memory efficiency: Yield one value at a time (great for large/infinite sequences).
 - Lazy evaluation: Values computed only when needed.
 - Faster startup: No need to generate entire collection upfront.

```python
# Regular function - loads everything in memory
def get_squares(n):
    return [x*x for x in range(n)]

# Generator - very memory efficient
def get_squares_gen(n):
    for x in range(n):
        yield x*x


print("=== Using Regular Function ===")
squares_list = get_squares(10)
print(squares_list)                    # Prints the full list
print(type(squares_list))              # <class 'list'>


print("\n=== Using Generator ===")
gen = get_squares_gen(10)

print(type(gen))                       # <class 'generator'>

# Printing one by one
print(next(gen))   # 0
print(next(gen))   # 1
print(next(gen))   # 4
print(next(gen))   # 9

# Using for loop with generator (Best way)
print("\nPrinting all using for loop:")
for sq in get_squares_gen(10):
    print(sq, end=" ")   # 0 1 4 9 16 25 36 49 64 81


# Useful for reading huge files line by line, infinite sequences, etc.
```

10. What is the difference between map(), reduce(), and filter() functions in Python?

 - map(): Applies a function to every item → returns iterator of transformed items.
 - filter(): Filters items based on a predicate (returns True/False) → returns iterator of items that pass.
 - reduce() (from functools): Reduces iterable to a single value by cumulatively applying a function.

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]

map_nums = list(map(lambda x: x*2, nums))          # [2,4,6,8,10]
filter_nums = list(filter(lambda x: x % 2 == 0, nums))  # [2,4]
reduce_nums = reduce(lambda x, y: x + y, nums)       # 15

print("Map :", map_nums)
print("Filter : ",filter_nums)
print("Reduce : ",reduce_nums)
```

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even_numbers(numbers):
    total = 0

    for num in numbers:
        if num % 2 == 0:
            total += num

    return total


# Example
nums = [10, 15, 20, 25, 30]

result = sum_even_numbers(nums)

print("Sum of even numbers:", result)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(number):

    if number <= 1:
        return False

    for i in range(2, int(number ** 0.5) + 1):

        if number % i == 0:
            return False

    return True



num = 97

if is_prime(num):
    print(num, "is a Prime Number")
else:
    print(num, "is not a Prime Number")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:

    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):

        if self.count >= self.terms:
            raise StopIteration

        value = self.a

        self.a, self.b = self.b, self.a + self.b

        self.count += 1

        return value


fib = FibonacciIterator(10)

for num in fib:
    print(num, end=" ")
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(exponent):

    for i in range(exponent + 1):
        yield 2 ** i



for value in powers_of_two(5):
    print(value)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

# Step 1: Create a sample file

with open("sample.txt", "w") as file:
    file.write("Hello\n")
    file.write("Python\n")
    file.write("Generator\n")
    file.write("Function\n")

# Step 2: Generator Function

def read_file(filename):

    with open(filename, "r") as file:

        for line in file:
            yield line.strip()

# Step 3: Read file using generator

for line in read_file("sample.txt"):
    print(line)
```
