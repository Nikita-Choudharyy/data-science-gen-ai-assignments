2. Explain the difference between mutable and immutable data types with examples?
-  MUTABLE DATA TYPES:-
   - Mutable data types can be changed after creation.
   - Value in the same memory location.
   - Mutable types are used when frequent updates/modification are needed.
   - Example-dict,list,set.
- IMMUTABLE DATA TYPES:-
   - Immutable data types cannot be changed location.
   - Any change creates a new object in memory.
   - Immutable types are used for fixed data to ensure safety and avoid accidental changes.
  - Example-int,float,string,tuple.

```python
#immutable example
a = 5
b = a
a = 15
print("Immutable change",(a,b)) #a changed, b same

#mutable example
x = [1,2]
y = x
x.append(3)
print("Mutable change",(x,y)) #both changed
```

4. Describe how dictionaries store data?
-  A dictionary stores data in key-value pairs, Each value has a key(name) that is used to access it.
- Syntax:- dict_name = { key : value, key : value, .... }
- Python converts each key into a hash value. This hash value helps Python decide where to store the data in memory.
- Dictionaries use a hash table. A hash table has many small slots(called buckets) to store data.
- The hash value of a key decides which bucket the key-value pair will go into. This makes searching very fast.
- If two keys get the same hash(collision), Python stores one of them in the next empty bucket so that the doctionary still works quickly.

14. How does the “in” keyword work differently for lists and dictionaries?
 - "in" in Lists:-
   - Check for values.
   - When you use "in" with a list, Python checks whether a value is present inside the list. It goes through the list elements one by one.
    - Python checks elements of the list:-
      - If the value exists- True
      - If the value doesn't exist- False

 -  "in" in Dictionaries:-
    - A dictionary stores data as key-value pairs. When we use "in" with a dictionary, Python checks only the keys, not the values.
    - To check values:- "ajay" in student.values()

```python
#"in" in list
nums = [10,20,30]
print(20 in nums)
print(40 in nums)

print('-'*30)

#"in" in Dictionaries
student = {"name":"ajay", "age":24}
print("name" in student)             #it will check for the key. If existS, it returnS true
print("ajay" in student)             #it will check for the key. Since "ajay" is not a key, it returns false
print("ajay" in student.values())    #correct way to check a value inside dictionary
```

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
 - Dictionaries are considered unordered because Python does not store their items in a fixed position like lists do.
 - The elements are placed based on hash values, not by index, so their order is not guaranteed to stay the same.
 - Because of this, you cannot access dictionary items by index(like dict[0]), only by using keys.
 - The order may change internally if you add or remove items, but this does not affect finding a value.
  - Data retrieval is still very fast(O(1)) because Python uses keys and hashing to directly reach the value, even though the order is not fixed.
