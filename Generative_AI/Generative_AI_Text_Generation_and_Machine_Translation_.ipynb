# Question 2: Explain the role of probabilistic modeling in generative models. How do these models differ from discriminative models?


# Role of Probabilistic Modeling in Generative Models

Probabilistic modeling plays several important roles in generative models:

### 1. Learns Data Distribution

It learns how the training data is distributed, allowing the model to understand patterns and relationships between features.



### 2. Generates New Data

Once the probability distribution is learned, the model can generate new text, images, audio, or other data that closely resembles the training data.


### 3. Handles Uncertainty

Real-world data often contains uncertainty and variation. Probabilistic models account for this uncertainty, making the generated outputs more realistic.



### 4. Improves Sampling

By sampling from the learned probability distribution, the model can create diverse and meaningful outputs instead of producing identical results every time.



### 5. Supports Missing Data Estimation

Probabilistic models can estimate missing or incomplete information based on the learned distribution, making them useful for data completion tasks.



# Difference Between Generative and Discriminative Models

| Feature | Generative Models | Discriminative Models |
|---------|-------------------|-----------------------|
| **Purpose** | Generate new data and model how data is created. | Classify or predict labels for existing data. |
| **Probability Learned** | Learns the **joint probability distribution** of data. | Learns the **conditional probability** of labels given the input. |
| **Output** | Generates new samples such as text, images, or audio. | Predicts class labels or numerical values. |
| **Main Objective** | Understand and reproduce the data distribution. | Separate or classify different classes accurately. |
| **Examples** | GAN, VAE, GPT | Logistic Regression, SVM, Random Forest |



# Question 3: What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?
-

# Autoencoder

An **Autoencoder (AE)** is a neural network that learns to compress input data into a smaller representation called the **latent vector** and then reconstruct the original input from this representation.

### Architecture of an Autoencoder

1. **Encoder**
   - Compresses the input text into a low-dimensional latent representation.

2. **Latent Space**
   - Stores the compressed representation of the input.

3. **Decoder**
   - Reconstructs the original text from the latent representation.

### Role in Text Generation

- Learns useful text features.
- Mainly reconstructs existing text.
- Does not generate diverse or completely new text effectively.



#  Variational Autoencoder (VAE)

A **Variational Autoencoder (VAE)** is an extension of the Autoencoder that learns a **probability distribution** over the latent space instead of a single fixed latent vector.

Instead of encoding each input into one point, a VAE encodes it as a **distribution (mean and variance)**. During generation, new latent vectors are sampled from this distribution and passed to the decoder to generate new text.

### Architecture of a VAE

1. **Encoder**
   - Learns the mean and variance of the latent distribution.

2. **Latent Distribution**
   - Represents the data as a probability distribution rather than a fixed vector.

3. **Sampling Layer**
   - Randomly samples latent vectors from the learned distribution.

4. **Decoder**
   - Generates new text from the sampled latent vector.

### Role in Text Generation

- Generates new and diverse sentences.
- Produces smoother latent representations.
- Suitable for creative text generation and language modeling.



# Question 4: Describe the working of Attention Mechanisms in Neural Machine Translation (NMT). Why are they critical?


# Working of the Attention Mechanism

The Attention Mechanism works through the following steps:

### Step 1: Encode the Input Sentence

The encoder processes each word in the source sentence and generates a sequence of hidden states. Each hidden state contains information about a particular word and its surrounding context.



### Step 2: Calculate Attention Scores

When generating each word in the translated sentence, the decoder compares its current hidden state with all encoder hidden states.

This comparison produces **attention scores**, which indicate how relevant each source word is to the current target word.



### Step 3: Compute Attention Weights

The attention scores are converted into **attention weights** using the **Softmax** function.

These weights represent the importance of each input word, and their values sum to 1.



### Step 4: Create the Context Vector

A **context vector** is computed as the weighted sum of all encoder hidden states using the attention weights.

This context vector contains the most relevant information needed to generate the next translated word.


### Step 5: Generate the Output Word

The decoder combines:

- Its current hidden state.
- The context vector.

Using this information, it predicts the next word in the translated sentence.

The same process is repeated until the complete translated sentence is generated.



# Why is Attention Critical?

The Attention Mechanism is essential because it addresses several limitations of traditional encoder-decoder models.

### 1. Handles Long Sentences

Instead of relying on a single fixed-length vector, attention allows the decoder to access all encoder hidden states, improving translations of long sentences.



### 2. Improves Translation Accuracy

The model focuses on the most relevant words while translating each target word, leading to more accurate and context-aware translations.



### 3. Captures Context Better

Attention helps the model understand relationships between words that are far apart in a sentence.



### 4. Reduces Information Loss

Important information from the source sentence is preserved because the decoder can refer back to any part of the input whenever needed.



### 5. Forms the Foundation of Modern NLP Models

Attention is the core idea behind **Transformer** architectures, which power modern Large Language Models (LLMs) such as **BERT**, **GPT**, and many advanced machine translation systems.

###Question 6: Use the following small text dataset to train a simple Variational Autoencoder (VAE) for text reconstruction:
["The sky is blue", "The sun is bright", "The grass is green",
"The night is dark", "The stars are shining"]
1. Preprocess the data (tokenize and pad the sequences)
2. Build a basic VAE model for text reconstruction
3. Train the model and show how it reconstructs or generates similar sentences.

```python
# Import required libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Lambda
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import backend as K

# ---------------------------------------------------
# Step 1: Dataset
# ---------------------------------------------------

texts = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

# ---------------------------------------------------
# Step 2: Text Preprocessing
# ---------------------------------------------------

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

max_length = max(len(seq) for seq in sequences)

X = pad_sequences(sequences,
                  maxlen=max_length,
                  padding="post")

vocab_size = len(tokenizer.word_index) + 1

print("Vocabulary Size:", vocab_size)
print("\nPadded Sequences:\n")
print(X)

# ---------------------------------------------------
# Step 3: Build Encoder
# ---------------------------------------------------

embedding_dim = 16
latent_dim = 8

inputs = Input(shape=(max_length,))

x = Embedding(vocab_size,
              embedding_dim)(inputs)

x = LSTM(32)(x)

z_mean = Dense(latent_dim)(x)
z_log_var = Dense(latent_dim)(x)

# Sampling Function
def sampling(args):
    z_mean, z_log_var = args

    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim))

    return z_mean + K.exp(0.5 * z_log_var) * epsilon

z = Lambda(sampling)([z_mean, z_log_var])

# ---------------------------------------------------
# Step 4: Decoder
# ---------------------------------------------------

decoder = Dense(max_length, activation="linear")

outputs = decoder(z)

vae = Model(inputs, outputs)

# ---------------------------------------------------
# Step 5: Compile Model
# ---------------------------------------------------

vae.compile(
    optimizer="adam",
    loss="mse"
)

# ---------------------------------------------------
# Step 6: Train Model
# ---------------------------------------------------

history = vae.fit(
    X,
    X.astype("float32"),
    epochs=100,
    batch_size=2,
    verbose=1
)

# ---------------------------------------------------
# Step 7: Reconstruction
# ---------------------------------------------------

reconstructed = vae.predict(X)

print("\nReconstructed Output:\n")

print(np.round(reconstructed, 2))
```

### Explanation

1. The text dataset is first tokenized using the Keras **Tokenizer**.
2. Each sentence is converted into numerical sequences.
3. The sequences are padded to the same length using **pad_sequences()**.
4. The encoder converts the input sequence into a latent representation by learning:
   - Mean (z_mean)
   - Variance (z_log_var)
5. A sampling layer generates a latent vector from this probability distribution.
6. The decoder reconstructs the original sequence from the latent vector.
7. After training, the model attempts to reconstruct sentences that are similar to the original inputs.

###Question 8: Implement a simple attention-based encoder-decoder model for English-to-Spanish translation using Tensorflow or PyTorch.


```python
# Import required libraries
import tensorflow as tf
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Attention, Concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# ----------------------------------------------------
# Sample Dataset
# ----------------------------------------------------

english_sentences = [
    "hello",
    "how are you",
    "good morning",
    "thank you",
    "i love you"
]

spanish_sentences = [
    "hola",
    "como estas",
    "buenos dias",
    "gracias",
    "te amo"
]

# ----------------------------------------------------
# Tokenization
# ----------------------------------------------------

eng_tokenizer = Tokenizer()
eng_tokenizer.fit_on_texts(english_sentences)

spa_tokenizer = Tokenizer()
spa_tokenizer.fit_on_texts(spanish_sentences)

encoder_input = eng_tokenizer.texts_to_sequences(english_sentences)
decoder_output = spa_tokenizer.texts_to_sequences(spanish_sentences)

max_eng_len = max(len(seq) for seq in encoder_input)
max_spa_len = max(len(seq) for seq in decoder_output)

encoder_input = pad_sequences(
    encoder_input,
    maxlen=max_eng_len,
    padding="post"
)

decoder_output = pad_sequences(
    decoder_output,
    maxlen=max_spa_len,
    padding="post"
)

# ----------------------------------------------------
# Build Encoder
# ----------------------------------------------------

embedding_dim = 32
lstm_units = 64

encoder_inputs = Input(shape=(max_eng_len,))

encoder_embedding = Embedding(
    input_dim=len(eng_tokenizer.word_index) + 1,
    output_dim=embedding_dim
)(encoder_inputs)

encoder_outputs, state_h, state_c = LSTM(
    lstm_units,
    return_sequences=True,
    return_state=True
)(encoder_embedding)

# ----------------------------------------------------
# Build Decoder
# ----------------------------------------------------

decoder_inputs = Input(shape=(max_spa_len,))

decoder_embedding = Embedding(
    input_dim=len(spa_tokenizer.word_index) + 1,
    output_dim=embedding_dim
)(decoder_inputs)

decoder_outputs, _, _ = LSTM(
    lstm_units,
    return_sequences=True,
    return_state=True
)(
    decoder_embedding,
    initial_state=[state_h, state_c]
)

# ----------------------------------------------------
# Attention Layer
# ----------------------------------------------------

attention = Attention()(
    [decoder_outputs, encoder_outputs]
)

context = Concatenate()(
    [decoder_outputs, attention]
)

# ----------------------------------------------------
# Output Layer
# ----------------------------------------------------

outputs = Dense(
    len(spa_tokenizer.word_index) + 1,
    activation="softmax"
)(context)

# ----------------------------------------------------
# Build Model
# ----------------------------------------------------

model = Model(
    [encoder_inputs, decoder_inputs],
    outputs
)

# Compile Model
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# ----------------------------------------------------
# Prepare Target Data
# ----------------------------------------------------

decoder_target = np.expand_dims(decoder_output, -1)

# ----------------------------------------------------
# Train Model
# ----------------------------------------------------

history = model.fit(
    [encoder_input, decoder_output],
    decoder_target,
    epochs=50,
    batch_size=2,
    verbose=1
)

# ----------------------------------------------------
# Model Summary
# ----------------------------------------------------

print("\nModel Summary:\n")
model.summary()
```

### Explanation

1. A small EnglishвҖ“Spanish dataset is created.
2. The sentences are tokenized using the Keras **Tokenizer**.
3. Both English and Spanish sequences are padded to the same length.
4. The **Encoder** uses an Embedding layer followed by an LSTM to encode the English sentence.
5. The **Decoder** uses another Embedding layer and LSTM to generate the Spanish translation.
6. An **Attention layer** computes the relationship between the encoder outputs and decoder outputs, allowing the decoder to focus on the most relevant words in the input sentence.
7. The context vector from the Attention layer is concatenated with the decoder output and passed through a Dense layer with a **Softmax** activation function to predict the translated words.
8. The model is trained using the **Adam optimizer** and **Sparse Categorical Crossentropy** loss.

###Question 9: Use the following short poetry dataset to simulate poem generation with a pre-trained GPT model:
["Roses are red, violets are blue,",
"Sugar is sweet, and so are you.",
"The moon glows bright in silent skies,",
"A bird sings where the soft wind sighs."]
- Using this dataset as a reference for poetic structure and language, generate a new 2-4
line poem using a pre-trained GPT model (such as GPT-2). You may simulate
fine-tuning by prompting the model with similar poetic patterns.

```python
# Import pipeline
from transformers import pipeline

# Load GPT-2 model
generator = pipeline(
    "text-generation",
    model="gpt2"
)

# Poetry dataset (Reference)
poems = [
    "Roses are red, violets are blue,",
    "Sugar is sweet, and so are you.",
    "The moon glows bright in silent skies,",
    "A bird sings where the soft wind sighs."
]

# Display reference dataset
print("Reference Poetry Dataset:\n")

for line in poems:
    print(line)

# Prompt used for generation
prompt = """
Write a short 2-4 line poem in a similar style.

Roses are red, violets are blue,
Sugar is sweet, and so are you.
The moon glows bright in silent skies,
A bird sings where the soft wind sighs.

Poem:
"""

print("\n" + "="*60)
print("Prompt Used:\n")
print(prompt)

# Generate poem
generated = generator(
    prompt,
    max_new_tokens=40,
    temperature=0.8,
    top_p=0.95,
    do_sample=True
)

# Display generated poem
print("\n" + "="*60)
print("Generated Poem:\n")
print(generated[0]["generated_text"])
```

### Explanation

1. The reference poetry dataset is provided to guide the writing style.
2. A prompt is created that includes the sample poems and instructs GPT-2 to generate a new 2вҖ“4 line poem.
3. The GPT-2 model is loaded using the Hugging Face `transformers` library.
4. The model generates a new poem based on the provided prompt.
5. Parameters such as **temperature** and **top_p** are used to encourage creative and diverse text generation.
