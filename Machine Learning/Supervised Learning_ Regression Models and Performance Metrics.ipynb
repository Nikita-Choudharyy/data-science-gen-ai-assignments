# 1.  What is Simple Linear Regression (SLR)? Explain its purpose.


**Simple Linear Regression (SLR)** is a supervised machine learning algorithm used to predict the value of one **dependent variable (Y)** using one **independent variable (X)**. It establishes a linear relationship between the two variables by fitting the best possible straight line through the data.

The mathematical equation of Simple Linear Regression is:

**Y = β₀ + β₁X + ε**

Where:

* **Y** = Dependent (Target) Variable
* **X** = Independent (Input) Variable
* **β₀** = Intercept (Value of Y when X = 0)
* **β₁** = Slope (Change in Y for every one-unit increase in X)
* **ε** = Error Term (Difference between actual and predicted values)

### Purpose of Simple Linear Regression

The main purposes of Simple Linear Regression are:

1. **Prediction:** It predicts the value of the dependent variable based on the independent variable.
2. **Relationship Analysis:** It helps understand how one variable affects another.
3. **Trend Identification:** It identifies whether the relationship between variables is positive or negative.
4. **Decision Making:** It supports business and research decisions by providing future predictions.



# 2. What are the key assumptions of Simple Linear Regression?

Simple Linear Regression (SLR) is based on several assumptions. These assumptions ensure that the regression model produces accurate predictions and reliable statistical results.

### 1. Linearity

There should be a **linear relationship** between the independent variable (X) and the dependent variable (Y). This means that changes in X should result in proportional changes in Y.

### 2. Independence of Errors

The residuals (prediction errors) should be **independent** of each other. The error associated with one observation should not influence the error of another observation.

### 3. Homoscedasticity

The variance of the residuals should remain **constant** across all values of the independent variable. If the spread of residuals changes significantly, the model suffers from **heteroscedasticity**.

### 4. Normality of Residuals

The residuals should be **normally distributed** with a mean of zero. This assumption is important for valid hypothesis testing and confidence intervals.

### 5. No Significant Outliers

The dataset should not contain extreme outliers because they can strongly influence the regression line and reduce the model's accuracy.

### 6. No Measurement Errors

The independent variable should be measured accurately. Large measurement errors can affect the reliability of the regression model.






# 3. Write the mathematical equation for a Simple Linear Regression model and explain each term.

The mathematical equation of a **Simple Linear Regression (SLR)** model is:

**Y = β₀ + β₁X + ε**

Where:

* **Y** = Dependent Variable (Target Variable)

  * The variable whose value is predicted by the model.

* **X** = Independent Variable (Predictor Variable)

  * The input variable used to predict the value of **Y**.

* **β₀ (Intercept)**

  * The value of **Y** when **X = 0**.
  * It represents the point where the regression line intersects the Y-axis.

* **β₁ (Slope or Regression Coefficient)**

  * It represents the change in **Y** for every one-unit increase in **X**.
  * A positive slope indicates that **Y** increases as **X** increases, while a negative slope indicates that **Y** decreases as **X** increases.

* **ε (Error Term or Residual)**

  * The difference between the actual value and the predicted value.
  * It represents the variation in **Y** that cannot be explained by the model.

## Example

Suppose the regression equation is:

**House Price = 50000 + 250 × Area**

Here:

* **House Price** is the dependent variable (**Y**).
* **Area** is the independent variable (**X**).
* **50000** is the intercept (**β₀**).
* **250** is the slope (**β₁**), which means the house price increases by **₹250** for every additional square foot of area.

For a house with an area of **1000 sq. ft.**:

**House Price = 50000 + (250 × 1000) = ₹300000**

# 5. What is the Method of Least Squares in Linear Regression?

The **Method of Least Squares** is a mathematical technique used in **Linear Regression** to find the **best-fit regression line** for a given dataset. The objective of this method is to minimize the difference between the actual values and the values predicted by the regression model.

The difference between the actual value and the predicted value is called the **residual (error)**.

### Mathematical Formula

The Method of Least Squares minimizes the **Sum of Squared Errors (SSE)**:

**SSE = Σ (Yᵢ − Ŷᵢ)²**

Where:

* **Σ** = Summation of all observations
* **Yᵢ** = Actual value
* **Ŷᵢ** = Predicted value from the regression model
* **(Yᵢ − Ŷᵢ)** = Residual (Prediction Error)

The errors are **squared** so that:

* Positive and negative errors do not cancel each other.
* Larger errors receive a greater penalty than smaller ones.

### Working of the Method

1. Collect the dataset.
2. Fit different possible regression lines.
3. Calculate the residual (error) for each data point.
4. Square each residual.
5. Add all squared residuals.
6. Select the regression line with the **smallest Sum of Squared Errors (SSE)**.



# 6. What is Logistic Regression? How does it differ from Linear Regression?
**Logistic Regression** is a supervised machine learning algorithm used for **classification problems**. It predicts the probability that an input belongs to a particular class, such as **Yes/No**, **0/1**, or **True/False**. Unlike Linear Regression, which predicts continuous values, Logistic Regression predicts **categorical outcomes**.

Logistic Regression uses the **Sigmoid (Logistic) Function** to convert the output into a probability value between **0 and 1**.

### Sigmoid Function

[
P(Y=1)=\frac{1}{1+e^{-z}}
]

Where:

* **P(Y=1)** = Probability of belonging to class 1
* **e** = Euler's constant (approximately 2.718)
* **z = β₀ + β₁X** = Linear combination of input features

If the predicted probability is greater than **0.5**, the observation is classified as **Class 1**; otherwise, it is classified as **Class 0**.



## Difference Between Linear Regression and Logistic Regression

| Feature       | Linear Regression          | Logistic Regression         |
| ------------- | -------------------------- | --------------------------- |
| Purpose       | Predicts continuous values | Predicts categorical values |
| Output        | Numerical value            | Probability (0 to 1)        |
| Problem Type  | Regression                 | Classification              |
| Output Range  | (-∞ to +∞)                 | 0 to 1                      |
| Function Used | Linear Equation            | Sigmoid (Logistic) Function |
| Example       | Predicting house price     | Predicting spam email       |



#7.  Name and briefly describe three common evaluation metrics for regression models.

Evaluation metrics are used to measure how well a regression model predicts the target values. They compare the actual values with the predicted values and help determine the accuracy of the model.

### 1. Mean Absolute Error (MAE)

**Mean Absolute Error (MAE)** is the average of the absolute differences between the actual values and the predicted values.

**Formula:**

[
MAE=\frac{1}{n}\sum_{i=1}^{n}|Y_i-\hat{Y}_i|
]

Where:

* **Yᵢ** = Actual value
* **Ŷᵢ** = Predicted value
* **n** = Number of observations

**Description:**

* Measures the average prediction error.
* Easy to understand because the error is expressed in the same units as the target variable.
* A lower MAE indicates better model performance.



### 2. Mean Squared Error (MSE)

**Mean Squared Error (MSE)** is the average of the squared differences between the actual and predicted values.

**Formula:**

[
MSE=\frac{1}{n}\sum_{i=1}^{n}(Y_i-\hat{Y}_i)^2
]

**Description:**

* Squares the errors, giving more importance to large errors.
* Sensitive to outliers.
* A lower MSE indicates a better regression model.



### 3. Root Mean Squared Error (RMSE)

**Root Mean Squared Error (RMSE)** is the square root of the Mean Squared Error.

**Formula:**

[
RMSE=\sqrt{\frac{1}{n}\sum_{i=1}^{n}(Y_i-\hat{Y}_i)^2}
]

**Description:**

* Measures the average prediction error in the same units as the target variable.
* Easier to interpret than MSE.
* Lower RMSE values indicate better model accuracy.



# 8. What is the purpose of the R-squared metric in regression analysis?

**R-squared (R²)**, also known as the **Coefficient of Determination**, is a statistical metric used to evaluate the performance of a regression model. It measures how well the independent variable(s) explain the variation in the dependent variable.

The value of **R²** ranges from **0 to 1**.

### Formula

[
R^2 = 1 - \frac{\sum (Y_i-\hat{Y}_i)^2}{\sum (Y_i-\bar{Y})^2}
]

Where:

* **Yᵢ** = Actual value
* **Ŷᵢ** = Predicted value
* **Ȳ** = Mean of the actual values
* **Σ(Yᵢ − Ŷᵢ)²** = Sum of Squared Errors (SSE)
* **Σ(Yᵢ − Ȳ)²** = Total Sum of Squares (SST)

### Purpose of R-squared

The main purposes of the R-squared metric are:

1. **Measures Model Performance**

   * It indicates how well the regression model fits the data.

2. **Explains Variability**

   * It shows the proportion of variation in the dependent variable that is explained by the independent variable(s).

3. **Compares Regression Models**

   * It helps compare different regression models. Generally, the model with the higher R² value provides a better fit.

4. **Evaluates Prediction Accuracy**

   * A higher R² value indicates that the model makes more accurate predictions.

### Interpretation of R-squared

| R² Value | Interpretation                                          |
| -------- | ------------------------------------------------------- |
| **0**    | The model explains none of the variation in the data.   |
| **0.50** | The model explains 50% of the variation.                |
| **0.75** | The model explains 75% of the variation.                |
| **1.00** | The model explains 100% of the variation (perfect fit). |



#9. Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

```python
# Import required libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 5, 4, 5])

# Create the Linear Regression model
model = LinearRegression()

# Train (fit) the model
model.fit(X, Y)

# Print the slope and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

# 10. How do you interpret the coefficients in a simple linear regression model?

In a **Simple Linear Regression** model, the relationship between the independent variable (**X**) and the dependent variable (**Y**) is represented by the equation:

**Y = β₀ + β₁X + ε**

Where:

* **Y** = Dependent Variable (Target Variable)
* **X** = Independent Variable (Predictor Variable)
* **β₀** = Intercept
* **β₁** = Slope (Regression Coefficient)
* **ε** = Error Term

The coefficients **β₀** (intercept) and **β₁** (slope) describe how the independent variable affects the dependent variable.

## 1. Intercept (β₀)

The **intercept** is the predicted value of the dependent variable (**Y**) when the independent variable (**X**) is equal to **0**.

**Example:**

Regression Equation:

**Salary = 20,000 + 5,000 × Experience**

Here:

* **Intercept (β₀) = 20,000**
* This means that when **Experience = 0 years**, the predicted salary is **₹20,000**.

## 2. Slope (β₁)

The **slope** represents the change in the dependent variable for every one-unit increase in the independent variable.

* If the slope is **positive**, the dependent variable increases as the independent variable increases.
* If the slope is **negative**, the dependent variable decreases as the independent variable increases.

**Example:**

Regression Equation:

**Salary = 20,000 + 5,000 × Experience**

Here:

* **Slope (β₁) = 5,000**
* This means that for every additional **1 year of experience**, the salary is expected to increase by **₹5,000**.

## Interpretation Example

Suppose the regression equation is:

**House Price = 50,000 + 250 × Area**

* **Intercept = 50,000**

  * When the house area is **0 square feet**, the predicted price is **₹50,000**.
* **Slope = 250**

  * For every additional **1 square foot**, the house price increases by **₹250**.
