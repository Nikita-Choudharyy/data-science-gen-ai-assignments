## Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

### What is Anomaly Detection?

Anomaly Detection is a machine learning technique used to identify **unusual, rare, or abnormal data points** that are significantly different from the majority of the data. These unusual observations are known as **anomalies** or **outliers**. Detecting such anomalies is important because they often indicate critical events such as fraud, system failures, cyberattacks, medical abnormalities, or manufacturing defects.

The main goal of anomaly detection is to distinguish **normal behavior** from **abnormal behavior** so that organizations can detect problems early and take appropriate action.

### Applications of Anomaly Detection
- Fraud detection in banking and finance
- Network intrusion and cybersecurity
- Medical diagnosis
- Manufacturing quality control
- Fault detection in IoT devices and sensors
- Predictive maintenance



## Types of Anomalies

### 1. Point Anomaly (Global Anomaly)

A **Point Anomaly** occurs when **a single data point is significantly different from the rest of the dataset**. It is the simplest and most common type of anomaly.

**Example:**
- In a dataset where employee salaries range from ₹30,000 to ₹80,000, one employee has a salary of ₹10,00,000. This unusually high salary is a point anomaly.
- A customer usually spends ₹500–₹2,000 on a credit card but suddenly makes a purchase of ₹5,00,000. This transaction is considered a point anomaly.

**Applications:**
- Credit card fraud detection
- Sensor fault detection
- Financial fraud detection



### 2. Contextual Anomaly (Conditional Anomaly)

A **Contextual Anomaly** occurs when a data point is considered abnormal **only in a specific context**, such as time, location, or season. The same value may be normal in one situation but abnormal in another.

**Example:**
- A temperature of **40°C** is normal during summer but highly unusual during winter.
- High electricity consumption during the daytime is expected, but the same consumption at **3:00 AM** may indicate abnormal activity.
- Heavy website traffic during a festival sale is normal, whereas the same traffic on a regular day may indicate a cyberattack.

**Applications:**
- Weather monitoring
- Energy consumption analysis
- Website traffic analysis
- Smart city applications


### 3. Collective Anomaly

A **Collective Anomaly** occurs when **a group of related data points forms an abnormal pattern**, even though each individual data point appears normal. The anomaly can only be identified by analyzing the entire sequence or group.

**Example:**
- Thousands of login requests received by a server within a few seconds may indicate a **DDoS attack**, even though each login request appears normal individually.
- A sequence of slightly abnormal heartbeats may indicate a heart disease, although each heartbeat alone looks normal.
- Multiple small changes in machine sensor readings together may signal an upcoming equipment failure.

**Applications:**
- Cybersecurity
- Healthcare monitoring
- Video surveillance
- Predictive maintenance



## Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

## 1. Isolation Forest

### Approach

Isolation Forest is an **ensemble-based anomaly detection algorithm**. It works on the idea that anomalies are **few in number and different from normal observations**, making them easier to isolate.

The algorithm creates multiple random decision trees by randomly selecting features and split values. Since anomalous data points are rare and different, they are isolated with fewer splits, resulting in shorter path lengths. Normal data points require more splits because they belong to dense regions.

### Suitable Use Cases

- Credit card fraud detection
- Network intrusion detection
- Financial transaction monitoring
- Manufacturing defect detection
- Large-scale anomaly detection problems


## 2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

### Approach

DBSCAN is a **density-based clustering algorithm** that groups together points located in dense regions while treating points in sparse regions as **noise (outliers)**.

The algorithm uses two parameters:

- **eps (ε):** Maximum distance between neighboring points.
- **min_samples:** Minimum number of neighboring points required to form a cluster.

Points that do not belong to any dense cluster are automatically identified as anomalies.

### Suitable Use Cases

- GPS and location data analysis
- Spatial data mining
- Image segmentation
- Customer segmentation
- Outlier detection in geographic datasets


## 3. Local Outlier Factor (LOF)

### Approach

Local Outlier Factor (LOF) is a **density-based anomaly detection algorithm** that compares the local density of each data point with the density of its neighboring points.

If a point has a significantly lower density than its neighbors, it is assigned a high LOF score and is considered an anomaly.

Unlike global methods, LOF detects anomalies based on the **local neighborhood**, making it effective when data density varies across different regions.

### Suitable Use Cases

- Medical diagnosis
- Fraud detection
- Sensor fault detection
- Network security
- Customer behavior analysis





## Question 3: What are the key components of a Time Series? Explain each with one example.

### What is a Time Series?

A **Time Series** is a sequence of observations recorded at regular intervals of time, such as daily, monthly, quarterly, or yearly. The main objective of time series analysis is to understand the underlying patterns in the data and use them for forecasting future values.

**Examples of time series data include:**
- Daily stock prices
- Monthly sales of a company
- Hourly temperature readings
- Annual rainfall
- Weekly website traffic

A time series is generally composed of **four key components**:



## 1. Trend (T)

### Definition

A **Trend** represents the long-term direction of the data over time. It shows whether the values are generally **increasing, decreasing, or remaining constant** over a long period.

A trend usually occurs because of factors such as population growth, technological advancements, inflation, or business expansion.

### Example

Suppose an e-commerce company records its annual sales for ten years. If the sales increase every year due to growing customer demand, this upward movement represents an **upward trend**.

**Real-life example:**
- Increasing number of internet users every year.
- Growth in online shopping over the last decade.

## 2. Seasonality (S)

### Definition

**Seasonality** refers to patterns that repeat at **regular and fixed intervals** such as daily, weekly, monthly, quarterly, or yearly.

Seasonal variations are caused by weather conditions, holidays, festivals, business cycles, or human behavior.

### Example

An ice cream shop experiences higher sales during the summer months and lower sales during winter every year. Since this pattern repeats annually, it is considered **seasonality**.

**Real-life example:**
- Higher retail sales during Diwali or Christmas.
- Increased electricity consumption during summer.
- Higher tourism during holiday seasons.



## 3. Cyclical Component (C)

### Definition

The **Cyclical Component** represents fluctuations that occur over a **long period**, usually lasting several years. Unlike seasonality, cyclical patterns **do not have a fixed duration** and are influenced by economic or business cycles.

### Example

The automobile industry experiences higher sales during periods of economic growth and lower sales during economic recessions. These fluctuations form business cycles.

**Real-life example:**
- Rise and fall of housing prices.
- Economic expansion followed by recession.
- Changes in stock market performance over several years.



## 4. Irregular or Random Component (R)

### Definition

The **Irregular Component**, also known as the **Random Component**, consists of unpredictable changes caused by unexpected events. These variations cannot be explained by trend, seasonality, or cyclical patterns.

### Example

A sudden pandemic, natural disaster, war, or financial crisis may cause an unexpected drop in airline passenger traffic. Such unpredictable changes are known as **random or irregular variations**.

**Real-life example:**
- COVID-19 affecting business sales.
- Floods disrupting agricultural production.
- Sudden stock market crash due to political events.



## Question 4: Define Stationarity in Time Series. How can you test and transform a non-stationary series into a stationary one?

### What is Stationarity in Time Series?

A **stationary time series** is a time series whose statistical properties remain **constant over time**. In other words, the data does not show long-term trends, changing variance, or changing seasonal patterns.

For a time series to be considered stationary, it should satisfy the following conditions:

- **Constant Mean:** The average value remains the same over time.
- **Constant Variance:** The variability (spread) of the data does not change over time.
- **Constant Covariance (Autocovariance):** The relationship between observations depends only on the time gap (lag) and not on the actual time.

Many forecasting models, such as **ARIMA**, assume that the input time series is stationary because stationary data produces more reliable and accurate predictions.


# How to Test Whether a Time Series is Stationary?

There are several methods to determine whether a time series is stationary.

## 1. Visual Inspection

The simplest method is to plot the time series and observe its behavior.

A stationary series generally has:
- Constant mean
- Constant variance
- No clear upward or downward trend
- No strong seasonal pattern

A non-stationary series usually shows:
- Increasing or decreasing trend
- Changing variance
- Seasonal fluctuations



## 2. Augmented Dickey-Fuller (ADF) Test

The **Augmented Dickey-Fuller (ADF) Test** is one of the most commonly used statistical tests for checking stationarity.

### Hypotheses

- **Null Hypothesis (H₀):** The time series is **non-stationary**.
- **Alternative Hypothesis (H₁):** The time series is **stationary**.

### Decision Rule

- If **p-value < 0.05**, reject the null hypothesis. The series is considered **stationary**.
- If **p-value ≥ 0.05**, fail to reject the null hypothesis. The series is **non-stationary**.



## 3. KPSS (Kwiatkowski–Phillips–Schmidt–Shin) Test

The KPSS test is another popular statistical test for stationarity.

### Hypotheses

- **Null Hypothesis (H₀):** The series is **stationary**.
- **Alternative Hypothesis (H₁):** The series is **non-stationary**.

Using both **ADF** and **KPSS** tests together provides a more reliable conclusion about stationarity.



# How to Transform a Non-Stationary Series into a Stationary Series?

If a time series is non-stationary, several transformation techniques can be applied.

## 1. Differencing

Differencing is the most widely used method to remove trends.

It calculates the difference between consecutive observations.

**Formula:**

**New Value = Current Value − Previous Value**

### Example

Sales:
```
100, 110, 120, 130, 140
```

After first-order differencing:
```
10, 10, 10, 10
```

The trend is removed, making the series more stationary.



## 2. Log Transformation

When the variance increases over time, a **log transformation** helps stabilize the variance.

### Example

Original values:
```
100, 500, 2000, 8000
```

After applying the logarithm, the large differences become much smaller, making the data more stable.

This technique is commonly used in financial and economic time series.



## 3. Moving Average

A moving average smooths short-term fluctuations and reduces random noise, making long-term patterns easier to analyze.

### Example

Instead of using individual daily sales values, the average of every 7 days is calculated to smooth the series.



## 4. Seasonal Differencing

If the data contains seasonality, seasonal differencing removes repeating seasonal effects.

### Example

Monthly sales:
```
January 2024 = 500
January 2025 = 520
```

Seasonal Difference:
```
520 − 500 = 20
```

This helps eliminate seasonal patterns.


## Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.


| Model | Structure | Handles Trend | Handles Seasonality | Uses External Variables | Suitable Applications |
|--------|-----------|---------------|---------------------|-------------------------|----------------------|
| **AR** | Uses previous observations (lag values). | No | No | No | Stock prices, sensor readings, stationary financial data. |
| **MA** | Uses previous forecast errors (residuals). | No | No | No | Short-term forecasting and error correction. |
| **ARIMA** | Combines AR, differencing, and MA. | Yes | No | No | Sales forecasting, demand prediction, economic data. |
| **SARIMA** | ARIMA with additional seasonal components. | Yes | Yes | No | Weather forecasting, retail sales, tourism demand. |
| **SARIMAX** | SARIMA with exogenous (external) variables. | Yes | Yes | Yes | Electricity demand, traffic forecasting, business forecasting with external factors. |



````python
# Question 6
# Load the AirPassengers dataset, plot the original time series, and decompose it into Trend, Seasonality, and Residual components.

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the AirPassengers dataset
data = get_rdataset("AirPassengers").data

# Create a datetime index
data["Month"] = pd.date_range(start="1949-01", periods=len(data), freq="MS")
data.set_index("Month", inplace=True)

# Rename the passenger column
data.rename(columns={"value": "Passengers"}, inplace=True)

# Plot the original time series
plt.figure(figsize=(12,5))
plt.plot(data["Passengers"], color="blue")
plt.title("AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.grid(True)
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(
    data["Passengers"],
    model="multiplicative",
    period=12
)

# Plot Trend, Seasonality, and Residuals
decomposition.plot()
plt.show()
````

````python
# Question 7
# Apply Isolation Forest on a numerical dataset (NYC Taxi Fare) to detect anomalies and visualize them on a 2D scatter plot.

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Load the NYC Taxi Fare dataset
# Replace the file name with your dataset path if needed
df = pd.read_csv("NYC_taxi_fare_data.csv")

# Keep only numerical columns and remove missing values
df = df[["fare_amount", "trip_distance"]].dropna()

# Train Isolation Forest
iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.02,
    random_state=42
)

# Predict anomalies
# 1 = Normal point
# -1 = Anomaly
df["Anomaly"] = iso_forest.fit_predict(df)

# Separate normal and anomaly points
normal = df[df["Anomaly"] == 1]
anomaly = df[df["Anomaly"] == -1]

# Print anomaly counts
print("Total Samples:", len(df))
print("Normal Points:", len(normal))
print("Anomaly Points:", len(anomaly))

# Visualize the anomalies
plt.figure(figsize=(10,6))

# Normal points
plt.scatter(
    normal["trip_distance"],
    normal["fare_amount"],
    c="blue",
    s=20,
    label="Normal"
)

# Anomaly points
plt.scatter(
    anomaly["trip_distance"],
    anomaly["fare_amount"],
    c="red",
    s=40,
    label="Anomaly"
)

plt.title("Isolation Forest Anomaly Detection (NYC Taxi Fare)")
plt.xlabel("Trip Distance")
plt.ylabel("Fare Amount")
plt.legend()
plt.grid(True)

plt.show()
````

````python
# Question 8
# Train a SARIMA model on the AirPassengers dataset, forecast the next 12 months, and visualize the results.

# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load the AirPassengers dataset
data = get_rdataset("AirPassengers").data

# Create a datetime index
data["Month"] = pd.date_range(
    start="1949-01-01",
    periods=len(data),
    freq="MS"
)

# Set the datetime column as index
data.set_index("Month", inplace=True)

# Rename the passenger column
data.rename(columns={"value": "Passengers"}, inplace=True)

# Set the monthly frequency (avoids warning)
data = data.asfreq("MS")

# Build the SARIMA model
model = SARIMAX(
    data["Passengers"],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

# Train the model
model_fit = model.fit(disp=False)

# Forecast the next 12 months
forecast = model_fit.forecast(steps=12)

# Create forecast dates
forecast_dates = pd.date_range(
    start=data.index[-1] + pd.DateOffset(months=1),
    periods=12,
    freq="MS"
)

# Create forecast DataFrame
forecast_df = pd.DataFrame({
    "Month": forecast_dates,
    "Forecast": forecast.values
})

# Print forecast values
print(forecast_df)

# Plot original data and forecast
plt.figure(figsize=(12, 6))

# Original time series
plt.plot(
    data.index,
    data["Passengers"],
    label="Original Data",
    color="blue"
)

# Forecast
plt.plot(
    forecast_dates,
    forecast,
    label="Forecast (Next 12 Months)",
    color="red",
    linestyle="--",
    linewidth=2
)

plt.title("SARIMA Forecast for AirPassengers Dataset")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.grid(True)

plt.show()
````

````python
# Question 9
# Apply Local Outlier Factor (LOF) on a numerical dataset to detect anomalies and visualize them using matplotlib.

# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import LocalOutlierFactor

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Use only the first two features for visualization
X_plot = X[:, :2]

# Apply Local Outlier Factor (LOF)
lof = LocalOutlierFactor(
    n_neighbors=20,
    contamination=0.05
)

# Predict anomalies
# 1 = Normal point
# -1 = Anomaly
labels = lof.fit_predict(X_plot)

# Separate normal and anomaly points
normal = X_plot[labels == 1]
anomaly = X_plot[labels == -1]

# Print results
print("Total Samples:", len(X_plot))
print("Normal Points:", len(normal))
print("Anomaly Points:", len(anomaly))

# Visualize the results
plt.figure(figsize=(10,6))

# Plot normal points
plt.scatter(
    normal[:, 0],
    normal[:, 1],
    color="blue",
    s=30,
    label="Normal"
)

# Plot anomalies
plt.scatter(
    anomaly[:, 0],
    anomaly[:, 1],
    color="red",
    s=60,
    marker="x",
    label="Anomaly"
)

plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.legend()
plt.grid(True)

plt.show()
````

# Question 10: You are working as a data scientist for a power grid monitoring company.Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage

## Real-Time Data Science Workflow

### 1. Data Collection

The first step is to collect energy consumption data from smart meters, power grid sensors, weather stations, and regional monitoring systems. Since the data is generated every 15 minutes, it is continuously stored in a streaming database for real-time analysis.



## 2. Data Preprocessing

Before building machine learning models, the data must be cleaned and prepared. The preprocessing steps include:

- Handling missing values.
- Removing duplicate records.
- Converting the timestamp into DateTime format.
- Sorting the data based on time.
- Creating time-based features such as hour, day, month, and weekday.
- Encoding categorical variables like region.
- Scaling numerical features if required.

These steps improve the quality of the data and increase model performance.



## 3. Detecting Anomalies in Streaming Data

For anomaly detection, **Isolation Forest** would be the preferred algorithm.

### Why Isolation Forest?

- Efficient for large datasets.
- Suitable for real-time streaming data.
- Detects unusual spikes and drops quickly.
- Does not require labeled anomaly data.
- Handles high-dimensional numerical data efficiently.

Although **Local Outlier Factor (LOF)** and **DBSCAN** can also detect anomalies, they are computationally more expensive and less suitable for continuous streaming applications.



## 4. Time Series Forecasting

For short-term electricity demand forecasting, **SARIMAX** would be the best choice.

### Why SARIMAX?

SARIMAX can model:

- Trend
- Seasonality
- External variables (weather conditions, holidays, region)

Since electricity consumption is influenced by weather and seasonal patterns, SARIMAX provides more accurate forecasts than ARIMA or SARIMA.


## 5. Model Validation and Monitoring

The forecasting model should be evaluated using:

- Mean Absolute Error (MAE)
- Root Mean Squared Error (RMSE)
- Mean Absolute Percentage Error (MAPE)

For anomaly detection, the following metrics can be used if labeled anomalies are available:

- Precision
- Recall
- F1-Score
- ROC-AUC Score

The models should be retrained periodically using newly collected data to maintain high prediction accuracy and adapt to changing consumption patterns.



## 6. Business Benefits

This solution provides several advantages to the power grid company:

- Predicts future electricity demand accurately.
- Detects abnormal spikes or drops in real time.
- Prevents equipment failures and power outages.
- Optimizes electricity generation and distribution.
- Reduces operational costs.
- Improves resource planning.
- Enhances customer satisfaction through reliable power supply.



## Conclusion

A real-time energy monitoring system combines **Isolation Forest** for anomaly detection and **SARIMAX** for demand forecasting. This combination enables early detection of abnormal consumption, accurate forecasting of future energy demand, efficient power distribution, and improved operational decision-making.

````python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.metrics import mean_absolute_error
from statsmodels.tsa.statespace.sarimax import SARIMAX

# ---------------------------------------
# Load Dataset
# ---------------------------------------
df = pd.read_csv("energy_usage.csv")

# Convert timestamp to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])

# Sort by timestamp
df = df.sort_values("timestamp")

# Set timestamp as index
df.set_index("timestamp", inplace=True)

# Set 15-minute frequency
df = df.asfreq("15min")

# ---------------------------------------
# Anomaly Detection using Isolation Forest
# ---------------------------------------

# Select numerical feature
X = df[["energy_usage"]]

# Train Isolation Forest
isolation_forest = IsolationForest(
    n_estimators=100,
    contamination=0.02,
    random_state=42
)

# Predict anomalies
df["Anomaly"] = isolation_forest.fit_predict(X)

# Print anomaly counts
print("Anomaly Detection Results:")
print(df["Anomaly"].value_counts())

# ---------------------------------------
# Time Series Forecasting using SARIMAX
# ---------------------------------------

# Create time series
ts = df["energy_usage"]

# Explicitly set frequency
ts.index.freq = "15min"

# Build SARIMAX model
model = SARIMAX(
    ts,
    order=(1,1,1),
    seasonal_order=(1,1,1,96),   # 96 intervals = 24 hours
    enforce_stationarity=False,
    enforce_invertibility=False
)

# Train model
model_fit = model.fit(disp=False)

# Forecast next 96 observations (24 hours)
forecast = model_fit.forecast(steps=96)

# Print forecast
print("\nForecast for Next 24 Hours:")
print(forecast)

# ---------------------------------------
# Model Evaluation (Example)
# ---------------------------------------

actual = ts.tail(96)

if len(actual) == len(forecast):
    mae = mean_absolute_error(actual, forecast)
    print("\nMean Absolute Error (MAE):", round(mae, 2))

# ---------------------------------------
# Plot Results
# ---------------------------------------

plt.figure(figsize=(14,6))

# Original data
plt.plot(
    ts.index,
    ts,
    label="Original Energy Usage",
    color="blue"
)

# Forecast dates
forecast_index = pd.date_range(
    start=ts.index[-1] + pd.Timedelta(minutes=15),
    periods=96,
    freq="15min"
)

# Forecast plot
plt.plot(
    forecast_index,
    forecast,
    color="red",
    linestyle="--",
    linewidth=2,
    label="Forecast (Next 24 Hours)"
)

plt.title("Energy Demand Forecast using SARIMAX")
plt.xlabel("Time")
plt.ylabel("Energy Usage")
plt.legend()
plt.grid(True)

plt.show()
````
