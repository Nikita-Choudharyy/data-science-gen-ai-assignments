#Regression & Its Evaluation

1.  What is Simple Linear Regression?
 - Simple Linear Regression (SLR) is a supervised machine learning algorithm used to model the relationship between one independent variable (X) and one dependent variable (Y). It predicts the value of the dependent variable by fitting the best possible straight line through the data points.

 - The relationship between the variables is represented by the equation:

    - Y=mX+c

      - Where:

         - Y = Dependent (target) variable
         - X = Independent (input) variable
         - m = Slope of the line (shows how much Y changes for a one-unit change in X)
         - c = Intercept (the value of Y when X = 0)

 - The objective of Simple Linear Regression is to find the line that minimizes the difference between the actual values and the predicted values. This is done using the Least Squares Method, which minimizes the sum of squared errors.

2. What are the key assumptions of Simple Linear Regression?
 - Simple Linear Regression relies on several assumptions to ensure that the model produces accurate and reliable predictions. If these assumptions are violated, the model's performance and interpretation may be affected.
 | Assumption                  | Description                                                 |
| --------------------------- | ----------------------------------------------------------- |
| **Linearity**               | A linear relationship exists between X and Y.               |
| **Independence of Errors**  | Residuals are independent of each other.                    |
| **Homoscedasticity**        | Residuals have constant variance across all values of X.    |
| **Normality of Residuals**  | Residuals follow a normal distribution.                     |
| **No Significant Outliers** | No extreme values that strongly affect the regression line. |
| **Accurate Measurement**    | Variables are measured accurately with minimal error.       |


3. What is heteroscedasticity, and why is it important to address in regression
models?
 - Heteroscedasticity is a condition in regression analysis where the variance of the residuals (errors) is not constant across all values of the independent variable. In other words, the spread of the prediction errors changes as the value of the predictor increases or decreases.

 - In a well-fitted linear regression model, the residuals should have constant variance, a property known as homoscedasticity. When this assumption is violated, the model exhibits heteroscedasticity.
 - Why is Heteroscedasticity Important to Address?
   1. Reduces Reliability of Statistical Tests : Heteroscedasticity produces incorrect standard errors, making hypothesis tests (t-tests and F-tests) unreliable.
   2. Leads to Incorrect Confidence Intervals : Confidence intervals may become too narrow or too wide, resulting in misleading conclusions.
   3. Decreases Prediction Accuracy : The regression model may produce less reliable predictions, especially for observations where the error variance is large.
   4. Violates Regression Assumptions : Simple Linear Regression assumes constant variance of residuals. Heteroscedasticity violates this assumption and reduces the validity of the model.

4. What is Multiple Linear Regression?
 - Multiple Linear Regression (MLR) is a supervised machine learning algorithm used to predict the value of a dependent variable (Y) using two or more independent variables (X₁, X₂, X₃, ...). It extends Simple Linear Regression, which uses only one independent variable.

 - The goal of Multiple Linear Regression is to determine how multiple factors together influence the target variable and to make accurate predictions.
 - Mathematical Equation : 0​+β1​X1​+β2​X2​+⋯+βn​Xn​+ε
    - Where:

       - Y = Dependent (target) variable
       - X₁, X₂, ..., Xₙ = Independent (predictor) variables
       - β₀ = Intercept (constant)
       - β₁, β₂, ..., βₙ = Regression coefficients representing the effect of each predictor
       - ε = Error term (residual)


5. What is polynomial regression, and how does it differ from linear
regression?
 - Polynomial Regression is a supervised machine learning algorithm used to model non-linear relationships between the independent variable(s) and the dependent variable. Although it is called "polynomial," it is still considered a type of linear regression because the model is linear with respect to its coefficients.

 - Instead of fitting a straight line, Polynomial Regression fits a curved line by adding polynomial terms to the regression equation. This allows the model to capture more complex patterns in the data.
  - Mathematical Equation

    - For a polynomial of degree 2 (Quadratic) : Y=β0​+β1​X+β2​X2+ε
    - For a polynomial of degree 3 (Cubic) : Y=β0​+β1​X+β2​X2+β3​X3+ε

         - Where:

             - Y = Dependent variable
             - X = Independent variable
             - β₀ = Intercept
             - β₁, β₂, β₃ = Regression coefficients
             - ε = Error term

 - Difference Between Linear Regression and Polynomial Regression
 | Feature        | Linear Regression           | Polynomial Regression                                       |
| -------------- | --------------------------- | ----------------------------------------------------------- |
| Relationship   | Linear                      | Non-linear (curved)                                         |
| Equation       | (Y = \beta_0 + \beta_1X)    | (Y = \beta_0 + \beta_1X + \beta_2X^2 + \cdots + \beta_nX^n) |
| Graph Shape    | Straight line               | Curve                                                       |
| Complexity     | Simple                      | More flexible and complex                                   |
| Best Used When | Data follows a linear trend | Data follows a curved or nonlinear trend                    |


6.Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
 -  X = [1, 2, 3, 4, 5]
 -  Y = [2.1, 4.3, 6.1, 7.9, 10.2]
 - Plot the regression line over the data points.


```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create the Linear Regression model
model = LinearRegression()

# Train (fit) the model
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Print model parameters
print("Intercept (b0):", model.intercept_)
print("Slope (b1):", model.coef_[0])

# Plot the original data points
plt.scatter(X, Y, color='blue', s=80, label='Actual Data')

# Plot the regression line
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')

# Add labels and title
plt.title("Simple Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

 - Conclusion : The Simple Linear Regression model successfully fits a straight line to the given data. The regression line shows a strong positive linear relationship between X and Y, indicating that as X increases, Y also increases approximately linearly.

7. Fit a Multiple Linear Regression model on this sample data:
 -  Area = [1200, 1500, 1800, 2000]
 -  Rooms = [2, 3, 3, 4]
 -  Price = [250000, 300000, 320000, 370000]
 - Check for multicollinearity using VIF and report the results.


```python
# Import required libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create the dataset
data = pd.DataFrame({
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
})

# Independent and dependent variables
X = data[['Area', 'Rooms']]
y = data['Price']

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Display coefficients
print("Intercept:", model.intercept_)
print("Coefficients:")
print("Area :", model.coef_[0])
print("Rooms:", model.coef_[1])

# Predict prices
predictions = model.predict(X)

# Display actual vs predicted values
result = pd.DataFrame({
    'Actual Price': y,
    'Predicted Price': predictions
})

print("\nPredictions:")
print(result)

# -----------------------------
# Calculate Variance Inflation Factor (VIF)
# -----------------------------
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [
    variance_inflation_factor(X.values, i)
    for i in range(X.shape[1])
]

print("\nVariance Inflation Factor (VIF):")
print(vif_data)
```


 - A Multiple Linear Regression model was successfully trained using Area and Rooms to predict Price. The VIF analysis indicates moderate to high multicollinearity between the independent variables, as both have VIF values greater than 5. Although the model can still make predictions, reducing multicollinearity (for example, by collecting more data or removing highly correlated features) may improve the stability and interpretability of the regression coefficients.

8. Implement polynomial regression on the following data:
 -  X = [1, 2, 3, 4, 5]
 -  Y = [2.2, 4.8, 7.5, 11.2, 14.7]
 - Fit a 2nd-degree polynomial and plot the resulting curve

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Sample Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Convert features into polynomial features (Degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Train the Polynomial Regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values
Y_pred = model.predict(X_poly)

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Generate smooth curve for plotting
X_new = np.linspace(1, 5, 100).reshape(-1, 1)
X_new_poly = poly.transform(X_new)
Y_new = model.predict(X_new_poly)

# Plot original data points
plt.scatter(X, Y, color='blue', s=80, label='Actual Data')

# Plot polynomial regression curve
plt.plot(X_new, Y_new, color='red', linewidth=2, label='Polynomial Regression (Degree = 2)')

# Labels and title
plt.title("Polynomial Regression (Degree = 2)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

 - Conclusion : The Polynomial Regression model of degree 2 successfully captures the nonlinear relationship between X and Y. Unlike Simple Linear Regression, which fits a straight line, Polynomial Regression fits a curved (quadratic) line, providing a better fit for datasets that exhibit nonlinear patterns.

9. Create a residuals plot for a regression model trained on this data:
 -  X = [10, 20, 30, 40, 50]
 -  Y = [15, 35, 40, 50, 65]
 - Assess heteroscedasticity by examining the spread of residuals.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Train the Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Print predictions and residuals
print("Predicted Values:")
print(Y_pred)

print("\nResiduals:")
print(residuals)

# Create Residual Plot
plt.figure(figsize=(7,5))
plt.scatter(Y_pred, residuals, color='blue', s=80)

# Horizontal line at zero
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)

# Labels and title
plt.title("Residual Plot")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.grid(True)

plt.show()
```

 - Assessment of Heteroscedasticity

    - The residuals are distributed around the zero line without a clear funnel-shaped or systematic pattern. Although the dataset is very small (only 5 observations), the spread of residuals appears relatively random.

       - There is no strong evidence of heteroscedasticity.
       - The residuals do not consistently increase or decrease as the predicted values increase.
       - A larger dataset would be required for a more reliable assessment of heteroscedasticity.

 - Conclusion : The Simple Linear Regression model was successfully trained, and the residual plot was created. Based on the residual distribution, there is no obvious sign of heteroscedasticity in this sample dataset. The residuals appear to be randomly scattered around zero, indicating that the assumption of constant variance (homoscedasticity) is reasonably satisfied for this small dataset.

10. Imagine you are a data scientist working for a real estate company. You

model. Explain the steps you would take to address these issues and ensure a robust
model.
 - As a data scientist working for a real estate company, my objective is to build an accurate and reliable regression model for predicting house prices. If the model shows heteroscedasticity and multicollinearity, I would follow the steps below to improve its performance and reliability.
 - Step 1: Understand the Data

              - Box-Cox transformation

         - Use Weighted Least Squares (WLS) when the error variance is not constant.
         - Use Robust Standard Errors to obtain more reliable statistical inference.

 - Step 3: Detect and Handle Multicollinearity
    - Detection
        - Calculate the Variance Inflation Factor (VIF) for each independent variable.
        - Examine the correlation matrix to identify highly correlated features.
    - Solutions
        - Remove highly correlated variables.
        - Combine correlated features into a single feature when appropriate.
        - Use feature selection techniques.
        - Apply dimensionality reduction methods such as Principal Component Analysis (PCA).
        - Consider regularization methods such as Ridge Regression or Lasso Regression, which reduce the impact of multicollinearity.

    - Scale numerical features if required.
    - Create new meaningful features, such as:
         - Price per square foot
         - Age of the property
         - Distance from schools, hospitals, and public transport


 - Step 5: Train and Validate the Model
    - Split the dataset into training and testing sets.
    - Train the regression model using the processed data.
    - Evaluate the model using:
         - R² Score
         - Mean Absolute Error (MAE)
         - Mean Squared Error (MSE)
         - Root Mean Squared Error (RMSE)
     - Use Cross-Validation to ensure the model generalizes well to unseen data.

 - Step 6: Verify Model Assumptions

    - After training the model, verify that:

        - The relationship between predictors and target is linear.
        - Residuals are randomly distributed around zero.
        - Residual variance is approximately constant (homoscedasticity).
        - Residuals are approximately normally distributed.
        - Multicollinearity has been reduced to acceptable levels (typically VIF < 5).

 - Expected Outcome

    - By addressing heteroscedasticity and multicollinearity:

         - The regression coefficients become more stable and interpretable.
         - Statistical tests and confidence intervals become more reliable.
         - Prediction accuracy improves.
         - The model generalizes better to new housing data.
         - Business decisions based on the model become more trustworthy.

I would identify multicollinearity using the Variance Inflation Factor (VIF) and address it through feature selection, dimensionality reduction, or regularization methods such as Ridge or Lasso Regression. Finally, I would validate the model using appropriate evaluation metrics and cross-validation.
