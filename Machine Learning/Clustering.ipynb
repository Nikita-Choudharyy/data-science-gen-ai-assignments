### Question 1: What is Unsupervised Learning in the context of Machine Learning?

**Answer:**

Unsupervised learning is a type of machine learning in which the model is trained using **unlabeled data**, meaning the dataset does not contain any target or output labels. The objective of unsupervised learning is to discover hidden patterns, structures, relationships, or groups within the data without any prior knowledge of the correct answers.

Instead of predicting a target value, the algorithm analyzes the similarities and differences among data points to organize them into meaningful clusters or identify underlying patterns. This makes unsupervised learning useful for exploring large datasets and gaining insights from data.



### Question 2: How does K-Means clustering algorithm work?

**Answer:**

K-Means is an unsupervised machine learning algorithm used to divide a dataset into **K distinct clusters** based on the similarity of data points. The goal is to group similar data points together while keeping different groups as separate as possible.

The K-Means algorithm works in the following steps:

1. **Choose the Number of Clusters (K):**
   - Select the number of clusters (K) that the dataset should be divided into.

2. **Initialize Centroids:**
   - Randomly choose K data points as the initial centroids (center of each cluster).

3. **Assign Data Points:**
   - Calculate the distance between each data point and every centroid.
   - Assign each data point to the cluster whose centroid is closest.

4. **Update Centroids:**
   - Compute the new centroid of each cluster by taking the mean of all data points assigned to that cluster.

5. **Repeat the Process:**
   - Repeat the assignment and centroid update steps until the centroids no longer change significantly or the maximum number of iterations is reached.

6. **Final Clusters:**
   - The algorithm stops when stable clusters are formed, and each data point belongs to the nearest cluster.



### Question 3: Explain the concept of a dendrogram in hierarchical clustering.

**Answer:**

A **dendrogram** is a tree-like diagram used in **Hierarchical Clustering** to visualize how data points or clusters are merged during the clustering process. It shows the hierarchical relationship between data points and helps understand the structure of the clusters.

In a dendrogram:

- Each **leaf (bottom node)** represents an individual data point.
- As the algorithm progresses, similar data points or clusters are merged together.
- Each **branch** represents the merging of two clusters.
- The **height** at which two branches merge indicates the distance or dissimilarity between those clusters. Clusters that merge at a lower height are more similar, while those merging at a higher height are less similar.

A dendrogram is also useful for deciding the **optimal number of clusters**. By drawing a horizontal line across the dendrogram, the number of branches it intersects represents the number of clusters formed.

**Answer:**

### Question 5: What are the advantages of DBSCAN over K-Means?

**Answer:**

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a density-based clustering algorithm that groups together data points that are closely packed while identifying points in low-density regions as outliers. Compared to K-Means, DBSCAN offers several important advantages.

### Advantages of DBSCAN over K-Means

1. **No Need to Specify the Number of Clusters**
   - Unlike K-Means, DBSCAN does not require the user to specify the number of clusters (K) before training.
   - It automatically determines the number of clusters based on the density of the data.

2. **Can Detect Clusters of Arbitrary Shapes**
   - K-Means works best with spherical or circular clusters.
   - DBSCAN can identify clusters of irregular shapes, such as curved or elongated clusters.

3. **Handles Outliers and Noise Effectively**
   - DBSCAN automatically identifies data points that do not belong to any cluster as **noise (outliers)**.
   - K-Means assigns every data point to a cluster, even if it is an outlier.

4. **Less Sensitive to Initialization**
   - K-Means depends on the initial placement of centroids, which can affect the final clustering results.
   - DBSCAN does not use centroids, making it more stable and consistent.

5. **Works Well with Unevenly Distributed Data**
   - DBSCAN performs well when clusters have different shapes and sizes, provided they have sufficient density.
   - K-Means may struggle with such datasets.



### Question 6: When would you use Silhouette Score in clustering?

**Answer:**

### When is Silhouette Score used?

1. **To Evaluate Clustering Quality**
   - It measures how similar a data point is to its own cluster and how different it is from other clusters.
   - A higher Silhouette Score indicates better-defined and well-separated clusters.

2. **To Choose the Optimal Number of Clusters (K)**
   - In algorithms like K-Means, different values of **K** can be tested.
   - The value of **K** with the highest average Silhouette Score is usually considered the best choice.

3. **To Compare Different Clustering Algorithms**
   - It can be used to compare the performance of algorithms such as K-Means, Hierarchical Clustering, and DBSCAN on the same dataset.

4. **To Check Cluster Separation**
   - It helps determine whether clusters are compact and clearly separated or overlapping with each other.



### Question 7: What are the limitations of Hierarchical Clustering?

**Answer:**


### Limitations of Hierarchical Clustering

1. **High Computational Cost**
   - Hierarchical Clustering requires calculating the distance between many pairs of data points.
   - As the size of the dataset increases, both the computation time and memory usage increase significantly.
   - Therefore, it is not suitable for very large datasets.

2. **Poor Scalability**
   - The algorithm works efficiently on small to medium-sized datasets but becomes slow and inefficient for large datasets containing thousands or millions of records.

3. **Sensitive to Noise and Outliers**
   - Outliers or noisy data can significantly affect the clustering process.
   - Once an outlier is merged into a cluster, it may reduce the overall quality of the clustering results.

4. **Irreversible Merging or Splitting**
   - In agglomerative hierarchical clustering, once two clusters are merged, they cannot be separated later.
   - Similarly, in divisive clustering, once a cluster is split, it cannot be merged back.
   - This lack of flexibility may lead to suboptimal clustering.

5. **Choice of Distance Metric and Linkage Method**
   - The final clustering result depends heavily on the selected distance measure (such as Euclidean or Manhattan distance) and linkage method (single, complete, average, or Ward linkage).
   - Different choices may produce different cluster structures.

6. **Difficulty in Selecting the Number of Clusters**
   - Although a dendrogram helps visualize the clustering process, determining the optimal number of clusters often requires manual interpretation, which can be subjective.



### Question 8: Why is feature scaling important in clustering algorithms like K-Means?

**Answer:**


1. **Ensures Equal Contribution of All Features**
   - Feature scaling brings all features to a similar range.
   - This prevents features with larger numerical values from having a greater influence on the clustering process.

2. **Improves Distance Calculations**
   - Since K-Means relies on distance measures, scaling ensures that distances are calculated fairly across all features.
   - This leads to more meaningful and accurate clusters.

3. **Produces Better Cluster Quality**
   - Scaled features help the algorithm identify natural groupings in the data instead of being influenced by differences in feature magnitudes.

4. **Faster Convergence**
   - Feature scaling helps the K-Means algorithm converge more quickly, reducing the number of iterations required to find the optimal cluster centroids.

5. **Improves Model Performance**
   - Properly scaled data often results in better clustering accuracy, improved cluster separation, and more reliable results.



### Question 9: How does DBSCAN identify noise points?

**Answer:**

### How DBSCAN Identifies Noise Points

1. **Find Neighboring Points**
   - For each data point, DBSCAN searches for all neighboring points that lie within the radius defined by **eps**.

2. **Identify Core Points**
   - If a data point has at least **MinPts** neighbors (including itself) within the eps radius, it is considered a **Core Point**.

3. **Identify Border Points**
   - A point that has fewer than **MinPts** neighbors but lies within the neighborhood of a Core Point is called a **Border Point**.
   - Border points are included in the nearest cluster.

4. **Identify Noise (Outlier) Points**
   - If a data point has fewer than **MinPts** neighbors and is **not connected to any Core Point**, it is classified as a **Noise Point (Outlier)**.
   - These points do not belong to any cluster because they are located in low-density regions.



### Question 10: Define inertia in the context of K-Means.

**Answer:**

**Inertia** is a performance metric used in the **K-Means clustering algorithm** to measure the quality of the clusters formed. It represents the **sum of the squared distances between each data point and the centroid of the cluster to which it belongs**. In other words, inertia measures how closely the data points are grouped around their respective cluster centers.

A **lower inertia value** indicates that the data points are closer to their cluster centroids, resulting in more compact and well-defined clusters. On the other hand, a **higher inertia value** means that the data points are more spread out within the clusters, indicating poorer clustering quality.





### Question 11: What is the Elbow Method in K-Means Clustering?

**Answer:**

The **Elbow Method** is a popular technique used in **K-Means clustering** to determine the **optimal number of clusters (K)** for a dataset. Since K-Means requires the value of **K** to be specified before training, the Elbow Method helps identify the most suitable number of clusters.

The method works by running the K-Means algorithm for different values of **K** (for example, from 1 to 10) and calculating the **inertia** for each value. Inertia is the sum of the squared distances between each data point and the centroid of its assigned cluster.

As the number of clusters increases, the inertia value decreases because data points become closer to their respective centroids. Initially, the decrease in inertia is significant, but after a certain point, adding more clusters results in only a small improvement. This point creates a bend or **"elbow"** in the graph.

The value of **K** at this elbow point is considered the optimal number of clusters because it provides a good balance between clustering accuracy and model complexity.



### Question 12: Describe the concept of "density" in DBSCAN.

**Answer:**


### How Density Works in DBSCAN

1. **Core Point**
   - A data point is called a **Core Point** if it has at least **MinPts** neighboring points (including itself) within the specified **eps** distance.
   - Core points are the foundation of a cluster because they lie in high-density regions.

2. **Border Point**
   - A **Border Point** has fewer than **MinPts** neighbors but lies within the eps radius of a Core Point.
   - Border points are assigned to the cluster of the nearby Core Point.

3. **Noise (Outlier) Point**
   - A point that has fewer than **MinPts** neighbors and is not within the neighborhood of any Core Point is considered a **Noise Point**.
   - These points do not belong to any cluster.



### Question 13: Can Hierarchical Clustering be used on categorical data?

**Answer:**

Yes, **Hierarchical Clustering can be used on categorical data**, but it cannot directly use distance measures such as **Euclidean distance**, which are designed for numerical data. Instead, it requires appropriate similarity or distance measures that are suitable for categorical variables.

For categorical datasets, distance measures such as **Hamming Distance**, **Jaccard Distance**, or **Gower Distance** can be used to calculate the similarity or dissimilarity between data points. Once the distance matrix is computed, Hierarchical Clustering can group similar observations into clusters and represent the results using a **dendrogram**.


### Question 14: What does a negative Silhouette Score indicate?

**Answer:**

A **Silhouette Score** measures how well a data point fits within its assigned cluster compared to other clusters. The score ranges from **-1 to +1**, where higher values indicate better clustering quality.

A **negative Silhouette Score** indicates that a data point is **closer to a neighboring cluster than to the cluster to which it has been assigned**. This suggests that the point may have been placed in the wrong cluster, resulting in poor clustering performance.



### Question 15: Explain the term "Linkage Criteria" in Hierarchical Clustering.

**Answer:**

**Linkage Criteria** is the method used in **Hierarchical Clustering** to determine **how the distance between two clusters is calculated** when deciding which clusters should be merged. Since hierarchical clustering combines clusters step by step, the linkage criterion plays a crucial role in determining the final cluster structure.

Different linkage methods calculate the distance between clusters in different ways, which can produce different clustering results even on the same dataset.



### Question 16: Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

**Answer:**


#### 1. Assumes Clusters Have Similar Sizes
- K-Means tries to create clusters of similar size.
- If one cluster is much larger than another, the algorithm may incorrectly divide the larger cluster or merge the smaller one with another cluster.

#### 2. Assumes Uniform Density
- K-Means performs best when all clusters have approximately the same density.
- If some clusters are very dense while others are sparse, the centroids may not accurately represent the actual cluster structure, leading to incorrect assignments.

#### 3. Uses Distance to the Centroid
- K-Means assigns each data point to the nearest centroid using **Euclidean distance**.
- In datasets with varying densities or irregular distributions, the nearest centroid may not represent the true cluster, causing misclassification.

#### 4. Cannot Handle Irregular Cluster Shapes
- K-Means works well for circular or spherical clusters.
- It struggles with clusters that are elongated, curved, or have complex shapes because centroid-based partitioning cannot capture these structures.

#### 5. Sensitive to Outliers
- Outliers can significantly shift the position of cluster centroids.
- This affects the assignment of nearby data points and reduces clustering accuracy.

### Example

Suppose a dataset contains:
- One **large and sparse** cluster.
- One **small and very dense** cluster.

K-Means may incorrectly split the large cluster into multiple clusters or merge the small dense cluster with the larger one because it attempts to minimize the distance to centroids rather than considering the density of data points.



### Question 17: What are the core parameters in DBSCAN, and how do they influence clustering?

**Answer:**

### Core Parameters in DBSCAN

#### 1. Epsilon (ε or eps)

- **Epsilon (eps)** is the maximum distance between two data points for them to be considered neighbors.
- It defines the radius of the neighborhood around each data point.

**Influence on Clustering:**
- **Small eps value:**
  - Creates smaller and tighter clusters.
  - More data points may be classified as noise because fewer neighbors fall within the radius.
- **Large eps value:**
  - Creates larger clusters.
  - Different clusters may merge together if the radius is too large.



#### 2. Minimum Points (MinPts)

- **MinPts** is the minimum number of data points required within the **eps** radius for a point to be considered a **Core Point**.
- It determines the minimum density required to form a cluster.

**Influence on Clustering:**
- **Small MinPts value:**
  - More clusters may be formed.
  - The algorithm becomes sensitive to noise and may create unnecessary clusters.
- **Large MinPts value:**
  - Requires denser regions to form clusters.
  - More points may be classified as noise if the density requirement is not met.


### Question 18: How does K-Means++ improve upon standard K-Means initialization?

**Answer:**

**K-Means++** is an improved initialization technique for the **K-Means clustering algorithm**. It was developed to overcome one of the main limitations of standard K-Means—the **random selection of initial centroids**. In standard K-Means, randomly chosen centroids can lead to poor clustering results, slow convergence, or convergence to a local optimum.

K-Means++ selects the initial centroids in a smarter way by ensuring that they are **well separated from each other**. This increases the likelihood of finding better clusters and improves the overall performance of the algorithm.



### Question 19: What is Agglomerative Clustering?

**Answer:**

**Agglomerative Clustering** is a type of **Hierarchical Clustering** and a popular **unsupervised machine learning algorithm** used to group similar data points into clusters. It follows a **bottom-up approach**, where each data point initially forms its own individual cluster. The algorithm then repeatedly merges the two most similar clusters until all data points belong to a single cluster or the desired number of clusters is obtained.

The similarity between clusters is determined using a **distance metric** (such as Euclidean or Manhattan distance) and a **linkage criterion** (such as Single Linkage, Complete Linkage, Average Linkage, or Ward Linkage).



### Question 20: What makes Silhouette Score a better metric than just inertia for model evaluation?

**Answer:**

Both **Inertia** and **Silhouette Score** are used to evaluate the performance of clustering algorithms such as **K-Means**, but the **Silhouette Score** provides a more complete and reliable measure of clustering quality.

**Inertia** measures the sum of the squared distances between each data point and the centroid of its assigned cluster. It only evaluates how compact the clusters are. As the number of clusters (**K**) increases, inertia always decreases, even if the clustering quality does not actually improve. Therefore, inertia alone cannot determine the best clustering model.

In contrast, the **Silhouette Score** evaluates both:
- **Cluster Cohesion:** How close data points are to other points in the same cluster.
- **Cluster Separation:** How well different clusters are separated from each other.

Because it considers both compactness and separation, the Silhouette Score provides a more balanced evaluation of clustering performance.

### Why Silhouette Score is Better than Inertia

1. **Measures Both Cohesion and Separation**
   - Inertia measures only the compactness of clusters.
   - Silhouette Score measures both how compact the clusters are and how well they are separated.

2. **Helps Choose the Optimal Number of Clusters**
   - Inertia always decreases as the number of clusters increases.
   - The Silhouette Score helps identify the value of **K** that produces the best overall clustering quality.

3. **Provides an Interpretable Score**
   - The Silhouette Score ranges from **-1 to +1**:
     - **+1:** Excellent clustering with well-separated clusters.
     - **0:** Overlapping clusters.
     - **-1:** Poor clustering where points may be assigned to the wrong cluster.

4. **Useful for Comparing Different Clustering Models**
   - It allows comparison of different clustering algorithms or different values of **K** using a single evaluation metric.



```python
# Question 21
# Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic dataset with 4 centers
X, y = make_blobs(
    n_samples=400,
    centers=4,
    cluster_std=1.0,
    random_state=42
)

# Apply K-Means Clustering
kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

# Fit the model and predict clusters
y_pred = kmeans.fit_predict(X)

# Get cluster centroids
centroids = kmeans.cluster_centers_

# Plot the clusters
plt.figure(figsize=(8,6))

plt.scatter(
    X[:,0],
    X[:,1],
    c=y_pred,
    cmap="viridis",
    s=50,
    label="Data Points"
)

# Plot centroids
plt.scatter(
    centroids[:,0],
    centroids[:,1],
    c="red",
    marker="X",
    s=250,
    label="Centroids"
)

plt.title("K-Means Clustering on Synthetic Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Question 24
# Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster.

# Import required libraries
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Load the Wine dataset
wine = load_wine()

# Features
X = wine.data

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means Clustering
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

# Fit the model and predict cluster labels
labels = kmeans.fit_predict(X_scaled)

# Count the number of samples in each cluster
unique, counts = np.unique(labels, return_counts=True)

# Print the cluster sizes
print("Cluster Sizes:")

for cluster, size in zip(unique, counts):
    print(f"Cluster {cluster}: {size} samples")
```

```python
# Question 28
# Load the Digits dataset, reduce it to 2D using PCA, and visualize clusters from K-Means.

# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Load the Digits dataset
digits = load_digits()

# Features
X = digits.data

# Reduce the data to 2 dimensions using PCA
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

# Apply K-Means Clustering
kmeans = KMeans(
    n_clusters=10,
    random_state=42,
    n_init=10
)

# Fit the model and predict cluster labels
labels = kmeans.fit_predict(X_pca)

# Plot the clusters
plt.figure(figsize=(8,6))

plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=labels,
    cmap="tab10",
    s=40
)

# Plot cluster centroids
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    c="black",
    marker="X",
    s=200,
    label="Centroids"
)

plt.title("K-Means Clustering on Digits Dataset (PCA - 2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Question 29
# Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display the scores as a bar chart.

# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic dataset
X, y = make_blobs(
    n_samples=500,
    centers=4,
    cluster_std=1.0,
    random_state=42
)

# Store silhouette scores
k_values = [2, 3, 4, 5]
scores = []

# Calculate silhouette score for each value of k
for k in k_values:
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Print the silhouette scores
print("Silhouette Scores:")
for k, score in zip(k_values, scores):
    print(f"k = {k}: {score:.4f}")

# Plot the bar chart
plt.figure(figsize=(8,5))

plt.bar(
    [str(k) for k in k_values],
    scores,
    color="skyblue",
    edgecolor="black"
)

plt.title("Silhouette Scores for Different Values of k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(axis="y")

plt.show()
```

```python
# Question 31
# Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with overlapping clusters
X, y = make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=2.5,   # Larger value creates overlapping clusters
    random_state=42
)

# Apply K-Means
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

kmeans.fit(X)

# Create a mesh grid for decision boundaries
h = 0.05

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, h),
    np.arange(y_min, y_max, h)
)

# Predict cluster labels for each point in the mesh
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries
plt.figure(figsize=(8,6))

plt.contourf(
    xx,
    yy,
    Z,
    alpha=0.3,
    cmap="viridis"
)

# Plot data points
plt.scatter(
    X[:, 0],
    X[:, 1],
    c=kmeans.labels_,
    cmap="viridis",
    s=40
)

# Plot centroids
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    c="red",
    marker="X",
    s=250,
    label="Centroids"
)

plt.title("K-Means Clustering with Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Question 34
# Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show the results in a line plot.

# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Calculate inertia for different values of K
k_values = range(2, 7)
inertia = []

for k in k_values:
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Print inertia values
print("Inertia Values:")
for k, value in zip(k_values, inertia):
    print(f"K = {k}: {value:.2f}")

# Plot the inertia values
plt.figure(figsize=(8,6))

plt.plot(
    k_values,
    inertia,
    marker='o',
    linewidth=2
)

plt.title("K-Means Inertia for Different Values of K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.xticks(k_values)
plt.grid(True)

plt.show()
```

```python
# Question 44
# Load the Iris dataset and plot the Silhouette Coefficient for each sample after K-Means clustering.

# Import required libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means Clustering
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

cluster_labels = kmeans.fit_predict(X_scaled)

# Compute silhouette scores
sample_silhouette_values = silhouette_samples(X_scaled, cluster_labels)
average_score = silhouette_score(X_scaled, cluster_labels)

# Print average silhouette score
print("Average Silhouette Score:", round(average_score, 4))

# Plot silhouette coefficients
plt.figure(figsize=(8, 6))

y_lower = 10

for i in range(3):
    ith_cluster_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_values.sort()

    size_cluster = ith_cluster_values.shape[0]
    y_upper = y_lower + size_cluster

    plt.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        ith_cluster_values,
        alpha=0.7
    )

    plt.text(-0.05, y_lower + 0.5 * size_cluster, str(i))
    y_lower = y_upper + 10

# Plot average silhouette score line
plt.axvline(
    x=average_score,
    color="red",
    linestyle="--",
    label="Average Score"
)

plt.title("Silhouette Coefficient for Each Sample")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Cluster")
plt.legend()
plt.grid(True)

plt.show()
```
